```python
import pandas as pd
```

```python
fortune_500_data = pd.read_csv('./Resources/use_this_fortune_500.csv')
fortune_500_data
```

```python
fortune_500_df = fortune_500_data[["name", "rank", "year", "industry", "sector", "headquarters_state", "headquarters_city", "market_value_mil","revenue_mil","profit_mil","employees"]]

fortune_500_df
```

```python
renamed_df = fortune_500_df.rename(columns={"market_value_mil":"market_value_millions", "revenue_mil":"revenue_millions", "profit_mil":"profit_millions"})
renamed_df
```

```python
renamed_df.info()
```

```python
final_df = renamed_df.loc[renamed_df["year"]>= 2015]
final_df.reset_index(drop=True,inplace=True)
final_df
```

```python
final_df["market_value_billions"] = round(final_df["market_value_millions"] / 1000, 1)
final_df["revenue_billions"] = round(final_df["revenue_millions"] / 1000, 1)
final_df["profit_billions"] = round(final_df["profit_millions"] /1000, 1)
final_df
```

```python
final_df.columns
```

```python
reorder_df = final_df[['name', 'rank', 'year', 'industry', 'sector', 'headquarters_state',
       'headquarters_city', 'employees', 'market_value_millions', 'market_value_billions', 'revenue_millions', 'revenue_billions',
       'profit_millions', 'profit_billions' 
        ]]
reorder_df
```

LAT & LONG

```python
lat_log_data = pd.read_csv('./Resources/fortune_coordinates_new.csv')
lat_log_data
```

```python
clean_lat_long = lat_log_data[["latitude", "longitude"]]
clean_lat_long
```

```python
merge_df = pd.concat([reorder_df, clean_lat_long], axis=1)
merge_df
```

```python
merge_df.columns
```

```python
reorder_df_2 = merge_df[['name', 'rank', 'year', 'industry', 'sector', 'headquarters_state',
       'headquarters_city','latitude', 'longitude','employees', 'market_value_millions',
       'market_value_billions', 'revenue_millions', 'revenue_billions',
       'profit_millions', 'profit_billions']]
reorder_df_2
```
