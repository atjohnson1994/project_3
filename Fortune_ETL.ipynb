```python
# Import dependencies
import pandas as pd
import numpy as np
import re
pd.set_option('max_colwidth', 400)
```

```python
# Read the data into a Pandas DataFrame
fortune500_df = pd.read_csv('./Resources/kaggle-fortune500-1996-2023.csv')
pd.set_option('display.max_columns',55)
fortune500_df .head()
```

```python
# Display columns
fortune500_df.columns
```

```python
# Filter the DF
filtered_df = fortune500_df[['Name', 'Rank', 'Industry', 'Headquarters State', 'Revenues', 'year',
       'Headquarters City', 'Revenue Percent Change', 'Profits',
       'Profits Percent Change','Profit Percent Sales','Sales','Profit as a % of Sales','Gained in Rank', 
       'Dropped in Rank', 'Newcomer', 'Profitable','Growth in Jobs',
       'Sector', 'Revenues ($M)', 'Profits ($M)','Employees', 
       'Global 500','Change in Rank (Full 1000)','Change in Rank (500 only)']]

filtered_df
```

```python
# Rename the columns to keep the convention consistent
col_rename_df = filtered_df.rename(columns = { 'Headquarters State':'Headquarters_State','year':'Year',
       'Headquarters City':'Headquarters_City','Revenue Percent Change':'Revenue_Percent_Change',
       'Profits Percent Change':'Profits_Percent_Change','Profit Percent Sales':'Profit_Percent_Sales','Profit as a % of Sales':'Profit_as_a_%_of_Sales','Gained in Rank':'Gained_in_Rank', 
       'Dropped in Rank':'Dropped_in_Rank','Growth in Jobs':'Growth_in_Jobs',
       'Revenues ($M)':'Revenues_($M)', 'Profits ($M)':'Profits_($M)',
       'Global 500':'Global_500','Change in Rank (Full 1000)':'Change_in_Rank_(Full 1000)','Change in Rank (500 only)':'Change_in_Rank_(500 only)'})

col_rename_df
```

```python
# Check the data types of the renamed columns
col_rename_df.dtypes
```
