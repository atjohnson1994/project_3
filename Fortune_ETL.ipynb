# Mapping data
- Only looks at top 500 companies of each year
- Updates the 'Headquarters_City' field from null to the correct value
- Sends an API call to gather coordinates from city names
- Merges the city coordinates
- Sends an API call to gather coordinates from state names (when the city information is unavailable)
- Merges the state coordinates

```python
# Reduce results to top 500 from each year
reduced = col_rename_df.loc[revenue_df["Rank"] <= 500]
reduced.reset_index(inplace=True, drop=True)
reduced.head(5)
```

```python
# Generate list of companies in the dataset
name_list = reduced['Name']
name_list.value_counts()
```

```python
# Extract the company names from company list
name_list = name_list.value_counts()
name_list = name_list.index
name_list
```

```python
# Generate list of cities associated with each company name
headquarter_city = []
business_name = []
for i in name_list:
    item = reduced['Headquarters_City'].loc[reduced['Name'] == i]

    headquarter_city.append(item)
    business_name.append(i)
```

```python
# Generate new dataframe for company names and associated cities
business_name_and_headquarters_city = pd.DataFrame({
    'Headquarters_City': headquarter_city,
    'Name': business_name
})
business_name_and_headquarters_city.head(5)
```

```python
# For loop that removes (NAs, duplicates, numerics) and extracts the city name
headquarter_city2 = []
for i in range(len(business_name_and_headquarters_city['Headquarters_City'])):
    check = business_name_and_headquarters_city['Headquarters_City'][i].dropna()
    check = check.drop_duplicates()
    check = str(check)
    pattern = r'[0-9]'
    check = re.sub(pattern, '', check)
    check = check.split('\n')
    if check[0] == 'Series([], Name: Headquarters_City, dtype: object)':
        check[0] = None
        headquarter_city2.append(check[0])
    else:
        headquarter_city2.append(check[0].strip())
    
```

```python
# Generate new dataframe for condensed city name
business_name_and_headquarters_city_2 = pd.DataFrame({
    'Headquarters_City': headquarter_city2,
    'Name': business_name
})
business_name_and_headquarters_city_2.head(5)
```

```python
# Update the main dataframe with new city info
col_rename_df_updated = reduced
col_rename_df_updated = col_rename_df_updated.merge(business_name_and_headquarters_city_2, how ='left', on='Name')
col_rename_df_updated = col_rename_df_updated.rename(columns={'Headquarters_City_y': 'Headquarters_City'})
col_rename_df_updated = col_rename_df_updated.drop(columns='Headquarters_City_x')

col_rename_df_updated.head(5)
```
